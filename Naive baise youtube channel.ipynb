```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cross_validation import train_test_split
from sklearn import naive_bayes
from sklearn.metrics import roc_auc_score
import numpy as np
```

```python
df = pd.read_csv("training.txt", sep='\t', names=["linkes","text"])
```

```python
df.head(3)
```

```python
stopset = set(stopwords.words("english"))
vectorizer = TfidfVectorizer(use_idf=True, lowercase=True, strip_accents="ascii", stop_words=stopset)
```

```python
vectorizer
```

```python
#in this case our dependent variable will ne likes as 0(didn't like the movie) and 1 (liked the movie)
y = df.linkes
```

```python
X = vectorizer.fit_transform(df.text)
```

```python
print(y.shape)
print(X.shape)
```

```python
#Test train split
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state=10)
```

```python
X_train
```

```python
clf = naive_bayes.MultinomialNB()
clf.fit(X_train, y_train)
```

```python
#testing model accuracy
roc_auc_score(y_test, clf.predict_proba(X_test)[:,1])
```

```python
movie_review_array = np.array(["bad ducking anxiety need blaze lol stonerprobs."])

movie_review_vector = vectorizer.transform(movie_review_array)

print(clf.predict(movie_review_vector))
```
